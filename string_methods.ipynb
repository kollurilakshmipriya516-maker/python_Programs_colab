```python
#upper()
a="welcome to python class"
print(a.upper())
```

```python
#lower()
a="WELCOME TO PYTHON CLASS"
print(a.lower())
```

```python
#capitalize()
a="welcome to python class"
print(a.capitalize())
```

```python
#title()
a="welcome to python class"
print(a.title())
```

```python
#strip()
name="  python  "
print(name.strip())
```

```python
#lstrip()
name="  python  "
print(name.lstrip())
```

```python
#rstrip()
name=" python "
print(name.rstrip())
```

```python
#find()
a="welcome to python class"
print(a.find("O"))
```

```python
#index()
a="welcome to python class"
print(a.index("o"))
```

```python
#replace()
a="Good morning"
print(a.replace("morning","Afternoon"))
```

```python
#split
a="Good morning"
print(a.split())
```

```python
#join()
a=("30", "06", "2005")
x="-".join(a)
print(x)
```

```python
#isalpha()
a="python"
print(a.isalpha())
```

```python
#startswith()
a="python"
print(a.startswith("py"))
```

```python
#endswith()
a="python"
print(a.endswith("py"))
print(a.endswith("on"))
```

```python
#isalnum
a="python12"
print(a.isalnum())
```

```python
#isdigit()
a="123"
print(a.isdigit())
```

```python
#isnumeric()
a="123"
print(a.isnumeric())
```
