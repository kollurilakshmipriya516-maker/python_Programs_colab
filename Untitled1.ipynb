```python
#Increase Student Marks by Bonus Marks
marks=int(input("Enter marks:"))
bonus=int(input("Enter bonus marks:"))
marks+=bonus
print("marks=", marks)
```

```python
#Reduce product price after discount
price=float(input("Enter product price:"))
discount=float(input("Enter discount:"))
price-=discount
print("Final price =", int(price))
```

```python
#Multiply employee bonus by 3
bonus=int(input("Enter employee bonus:"))
bonus*=3
print("Updated bonus=",bonus)
```

```python
#Divide pizza slices among friends
slices=int(input("Enter total pizza slices:"))
friends=int(input("Enter number of friends:"))
slices/=friends
print("Slices per friend=",slices)
```

```python
#Find remainder after division
num=int(input("Enter number:"))
divisor=int(input("Enter divisor:"))
num%=divisor
print("Remainder=",num)
```

```python
#Create bank deposit program using +=
balance=float(input("Enter current balance:"))
deposit=float(input("Enter deposit amount:"))
balance+=deposit
print("Total balance=",balance)
```

```python
#Create electricity bill reduction using -=
bill=float(input("Enter electricity bill:"))
subsidy=float(input("Enter subsidy amount:"))
bill-=subsidy
print("Bill after reduction=",bill)
```

```python
#Find cube using **=
num=int(input("Enter a number: "))
num**=3
print("Cube=",num)
```

```python
#Create score update system for game
score=int(input("Enter current score:"))
points=int(input("Enter earned points:"))
score+=points
print("Updated score =",score)
```

```python
#Use all assignment operators in one program
num=int(input("Enter a number: "))
num+=5
print("After += 5 :",num)
num-=2
print("After -= 2 :",num)
num*=3
print("After *= 3 :",num)
num/=2
print("After /= 2 :",num)
num%=4
print("After %= 4 :",num)
num**=2
print("After **= 2 :",num)
num//=2
print("After //= 2 :",num)
```
