```python
"""
13. Leap Year Checker
Check whether a year is leap year.
"""
year=int(input("Enter year: "))
if(year%400==0) or (year%4==0 and year%100!=0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
"""
14. Character Checker
Check whether character is:

Alphabet
Digit
Special Character
"""
ch=input("Enter a character:")
if ch.isalpha():
    print("Alphabet")
elif ch.isdigit():
    print("Digit")
else:
    print("Special Character")
```

```python
"""
15. Vowel or Consonant
Check whether entered character is vowel or consonant.
"""
ch=input("Enter a character: ").lower()
if ch in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```

```python
"""
16. Simple Calculator

Take:
two numbers
operator (+, -, *, /)

Perform operation using conditions.
"""
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+,-,*,/): ")
if op == "+":
    print("Result =", int(a + b))
elif op == "-":
    print("Result =", int(a - b))
elif op == "*":
    print("Result =", int(a * b))
elif op == "/":
    print("Result =", a / b)
else:
    print("Invalid Operator")
```
